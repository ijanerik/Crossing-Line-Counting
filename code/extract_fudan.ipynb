```python
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image

%matplotlib inline

import datasets.fudan as fudan
```

```python
frame = fudan.load_video('data/Fudan/train_data/68').get_frames()[0]
```

```python
density = frame.get_density()
```

```python
cc_img = Image.fromarray(density * 255.0 / density.max())
cc_img = cc_img.convert("L")
cc_img.save('density.jpeg')
imshow(np.asarray(cc_img))
```
